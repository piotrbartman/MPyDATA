```python
from PyMPDATA_examples.Smolarkiewicz_2006_Figs_3_4_10_11_12.simulation import Simulation
from PyMPDATA_examples.Smolarkiewicz_2006_Figs_3_4_10_11_12.setup import Setup
from PyMPDATA import Options
import matplotlib.pyplot as plt
```

```python
def plot(simulation):
    psi_0 = simulation.state
    simulation.run()
    psi_T = simulation.state
    plt.plot(psi_0)
    plt.plot(psi_T)
```

```python
plot(Simulation(Setup("cosine"), Options(n_iters=1)))
```

```python
plot(Simulation(Setup("cosine"), Options(n_iters=2)))
```

```python
plot(Simulation(Setup("cosine"), Options(infinite_gauge=True, n_iters=2)))
```

```python
plot(Simulation(Setup("rect"), Options(infinite_gauge=True, n_iters=2)))
```

```python
plot(Simulation(Setup("rect"), Options(n_iters=2, infinite_gauge=True, flux_corrected_transport=True)))
```

```python
plot(Simulation(Setup("rect"), Options(n_iters=3, infinite_gauge=True, flux_corrected_transport=True)))
```
